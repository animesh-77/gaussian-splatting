# DFT on 2D image 

```python
# gratings.py
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
print(X.shape, Y.shape)

wavelength = 200
grating = np.sin(2 * np.pi * (X) / wavelength)
# grating = np.sin(2 * np.pi * (X-Y) / wavelength)
plt.set_cmap("gray")
print(grating.shape,)
plt.axis('on')
plt.imshow(grating)

# Remove axis and labels
plt.axis('off')

# Save the image as PNG
plt.savefig('grating.png', bbox_inches='tight', pad_inches=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
y= np.arange(-200,200,1)
X, Y = np.meshgrid(x, x)
wavelength = 200
angle = 0
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength
)
plt.set_cmap("gray")
plt.subplot(121)
plt.imshow(grating)
# Calculate Fourier transform of grating
ft = np.fft.ifftshift(grating)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)
plt.subplot(122)
plt.imshow(abs(ft))
plt.xlim([480, 520])
plt.ylim([480, 520])  # Note, order is reversed for y
plt.show()
```
